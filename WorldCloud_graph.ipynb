```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
from wordcloud import WordCloud, STOPWORDS 
```

```python
text = "AI.txt"
image_mask='AI.png'
```

```python
def WC_Graph(TEXT,mask):
    WC_Graph='WC_'+mask
    text = open(TEXT).read()
    image_mask = np.array(Image.open(mask))

    stopwords = set(STOPWORDS)
    #stopwords.add("said")

    wc = WordCloud(background_color="white", max_words=1000, mask=image_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

    wc.generate(text)
    wc.to_file(WC_Graph)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    
```

```python
WC_Graph(text,image_mask)
```
